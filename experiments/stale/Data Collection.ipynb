```python
# number of trials to simulate
N = 200
record = True
options = SUMOOptions()
options.nogui = True
options.model_path = os.path.abspath(os.path.join(os.pardir)) + "/sumo-files"
options.model_name = "lanechange"
df_dict = defaultdict(lambda: defaultdict(list))
dt = 0.1

for i in range(4, 5):
    trial = i
    tag = "_" + str(trial)
    scenario = generateDataFiles(i, record)
    trial = 'trial_' + str(i)
    print("computing " + trial)
    sumo = SUMOExperienceSource(options)
    state_trace = []
    p = 1
    step = 0
    while p:
        sys.stdout.flush()
        p = sumo.single_step()
        step += 1
        if p and step < 700:
            # get accel, omega features
            p = updateCurrentSnapshot(p, dt)
            state_trace.append(p)
        else:
            p = None
            if not sumo.sumo_alive:
                print("the sumo wrestler died at trial :" + trial)
            else:
                print("No cars or 1 car on the road")
        
    sys.stdout.flush()
    
    state_trace_adjacent = [updateSumostateGetAdjacentVeh(s) for s in state_trace]
    collision_filename =  os.path.abspath(os.path.join( os.pardir, 'data/errorlogs/error-log')) + tag + ".txt" 
    collision_dict = getCollisionDict(collision_filename)
    df = getDataFrame(state_trace_adjacent, trial, collision_dict, dt)
    
    df_dict[trial] = df
    sumo.close()
# sumo.close()

print("Data collection completed")
data_df = pd.concat(df_dict)
data_df.to_pickle( os.path.abspath(os.path.join( os.pardir, 'data/pickles/lanechange.pkl')))
```
